```python
import numpy as np
import query_arxiv as qa
import datetime

start = datetime.datetime.strptime('2024-09-01-03', "%Y-%m-%d-%H")
end = datetime.datetime.strptime('2024-10-15-18', "%Y-%m-%d-%H")

categories = ['astro-ph.GA', 'astro-ph.CO']
members = ['Ho Seong Hwang', 'Hyein Yoon', 'Jeong Hwan Lee', 'Daeun Jeong',
           'Dongkok Kim', 'Gain Lee', 'Hyeonguk Bahk', 'Sang Hyeok Im',
           'Sang Hyeon Han', 'Minseong Kwon', 'Yigon Kim', 'Wooseok Kang',
           'Jiwon Jang', 'Soojin Kim', 'Yonguk Cho', 'Tae Wan Kim', 'Marcie Mun']

members_abbr = []
members_more_abbr = []
for name in members:
    name_components = name.split()
    first_name_components = np.array(name_components)[:-1]
    name_abbr = ' '.join([nm[0]+'.' for nm in first_name_components] + [name_components[-1]])
    members_abbr.append(name_abbr)
    members_more_abbr.append(' '.join([name_components[0][0]+'.', name_components[-1]]))
```

```python

feed = qa.query_arxiv(categories, start, end, max_results=9000)
```

```python
feed_members = []
for entry in feed.entries:
    in_members = np.isin(members, [auth['name'] for auth in entry.authors])
    in_members_abbr = np.isin(members_abbr, [auth['name'] for auth in entry.authors])
    # in_members_more_abbr = np.isin(members_more_abbr, [auth['name'] for auth in entry.authors])
    in_members = in_members | in_members_abbr #| in_members_more_abbr
    if np.any(in_members):
        feed_members.append(entry)
        # print(np.array(members)[in_members], qa.make_markdown_entry(entry))
        print(qa.make_markdown_entry(entry))
```
